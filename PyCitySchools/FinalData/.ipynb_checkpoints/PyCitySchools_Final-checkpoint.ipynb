```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Number of schools is in school_data. Count the number of school names in the school data and output as .count().
school_number = school_data['school_name'].count()
school_number
```

```python
#Look at school_data format to use these information to extract data.
school_data.head(3)
```

```python
#Calculate total students from the school_data giving size of the school using sum(). Simpler than trying to use all data.
total_students = school_data['size'].sum()
total_students
```

```python
#Again, school_data gives the budget for each school, so sum() of the budget column gives the total budget for the district.
total_budget = school_data['budget'].sum()
total_budget
```

```python
#Look at student_data format to use these information for extracting data.
student_data.head(3)
```

```python
math_average = student_data['math_score'].mean()
math_average
```

```python
reading_average = student_data['reading_score'].mean()
reading_average
```

```python
passing_rate = (math_average + reading_average)/2
passing_rate
```

```python
passing_math = student_data.loc[student_data['math_score'] >= 70].count() / total_students
math_pass = passing_math['math_score'] * 100
```

```python
passing_reading = student_data.loc[student_data['reading_score'] >= 70].count() / total_students
reading_pass = passing_reading['reading_score'] * 100
```

```python
district_data = pd.DataFrame({'Number of Schools': [school_number], 'Number of Students': [total_students], 'Total Budget': [total_budget], 'Average Math Score': [math_average], 'Average Reading Score': [reading_average], 'Passing Rate': [passing_rate], 'Percent Pass Math': [math_pass], 'Percent Pass Reading': [reading_pass]})
```

```python
#dtypes on the district data indicates the Total Budget is an integer, not a float. Try to use astype to convert.
district_data.astype({'Total Budget': 'float'}).dtypes
```

```python
#district_data['Total Budget'] = district_data['Total Budget'].map("${:.2f}".format)
pd.options.display.float_format = '{:,.2f}'.format
district_data['Total Budget'] = district_data['Total Budget'].map("${:.2f}".format)
district_data
```

```python
#Save this information as a .csv file
district_data.to_csv('DistrictData.csv', index=False, header=True)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Look at school_data_complete to see how to extract data for final analyses.
school_data_complete.head(3)
```

```python
#Index school_data_complete on 'school_name' to enable indexing by school for subsequent analyses.
school_data_indexed = school_data_complete.set_index('school_name')
school_data_indexed.head(3)
```

```python
#Found a groupby approach to calculate means for reading and math. Also gets the SchoolID, size, and budget information.
#budget/size will give the average budget per student.
#Use these data to calculate the percent passing for math and reading, then add columns for these data. 
#School ID, size, and budget are all single values, so .mean() doesn't change anything on these values.
#See if I can use this groupby dataframe to merge with the school_data dataframe to link all data into one dataframe.
school_data_grouped = school_data_indexed.groupby('school_name')
school_data_grouped_average = school_data_grouped.mean()
school_data_grouped_average.head(3)
```

```python
#Calculate the average per student budget. Place in new column.
school_data_grouped_average['Budget per Student'] = school_data_grouped_average['budget'] / school_data_grouped_average['size']
school_data_grouped_average.head(3)
```

```python
#Calculate % passing for math and reading, then place in new columns. I can't use the grouped data because that
#has already combined information from students. I need individual student information. Probably need to use indexed
#data before grouping, school_data_indexed.
#Calculate percent passing by first extracting all students with math and reading above 70. Divide this by the total
#numbers of students in each school, the size column.
mathpassing = school_data_indexed.loc[school_data_indexed['math_score'] >= 70]
totalmathpassing = mathpassing.groupby('school_name')
```

```python
#Try to make a new column using the total of passing math scores in totalpassingmath dataframe, adding this to
#school_data_grouped_average dataframe.
school_data_grouped_average['Passed Math'] = totalmathpassing['size'].count()
school_data_grouped_average.head(3)
```

```python
#Repeat process from math with reading.
readpassing = school_data_indexed.loc[school_data_indexed['reading_score'] >= 70]
totalreadpassing = readpassing.groupby('school_name')
```

```python
school_data_grouped_average['Passed Reading'] = totalreadpassing['size'].count()
school_data_grouped_average.head(3)
```

```python
school_data_grouped_average['Percent Math Pass'] = (school_data_grouped_average['Passed Math'] / school_data_grouped_average['size'])*100
school_data_grouped_average.head(3)
```

```python
school_data_grouped_average['Percent Read Pass'] = (school_data_grouped_average['Passed Reading'] / school_data_grouped_average['size'])*100
school_data_grouped_average.head(3)
```

```python
school_data_grouped_average['Passing Rate'] = (school_data_grouped_average['Percent Math Pass'] + school_data_grouped_average['Percent Read Pass']) / 2
school_data_grouped_average.head(3)
```

```python
compiled_school_data = school_data_grouped_average[['size', 'budget', 'Budget per Student', 'Passed Math', 'Passed Reading', 'Percent Math Pass', 'Percent Read Pass', 'Passing Rate' ]]
compiled_school_data.head(3)
```

```python
#Organize headers into more informative names.
compiled_school_data = compiled_school_data.rename(columns = {'school_name': 'School', 'size': '# of Students', 'budget': 'Budget', 'Percent Math Pass': '% Pass Math', 'Percent Read Pass': '% Pass Reading', 'Passing Rate': '% Passing'})
compiled_school_data.head(3)
```

```python
#Original school_data has the information on charter versus district type of school. I can append this in a new column
#to these compiled_school_data. I will set a new index on school_data for school_name, then use this to maintain the order
#for merging.
original_school_list = school_data.set_index('school_name')
```

```python
#Add a column for this type information. One Confusing Thing!!! This adding of compiled_school_data School Type column
#used the original_school_list 'type' column, but mapped correctly to the actual schools! I suspect Pandas may keep track
#of the school names or, perhaps, if I used the index as the school name!
compiled_school_data['School Type'] = original_school_list['type']
compiled_school_data.head(3)
```

```python
#Now have to add the average scores for each school. Have this data in school_data_grouped_average dataframe in reading_score and math_score
#columns. Make new columns for compiled_school_data
compiled_school_data['Reading Average'] = school_data_grouped_average['reading_score']
compiled_school_data['Math Average'] = school_data_grouped_average['math_score']
compiled_school_data.head(3)
```

```python
#Reorganize into a final formatting for presentation and sort on % Passing.
organized_school_data = compiled_school_data[['Budget', 'Budget per Student', '# of Students', 'Math Average', 'Reading Average', '% Pass Math', '% Pass Reading', '% Passing', 'School Type']]
sorted_final_school_data = organized_school_data.sort_values('% Passing', ascending=False)
sorted_final_school_data['Budget'] = sorted_final_school_data['Budget'].map("${:.2f}".format)
sorted_final_school_data
```

```python
#Save sorted_final_school_data to a .csv file.
sorted_final_school_data.to_csv('SortedFinalSchoolData.csv', index=True, header=True)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:

  * Average Reading Score

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#Probably the best datafram to use for this would be the sorted_final_school_data. This has the average math, reading, passing% and overall passing%
#data. I will try using this to bin the data based on spending, then output information on the scores and % passing data.
sorted_final_school_data['Spending Per Student'] = pd.cut(sorted_final_school_data['Budget per Student'], spending_bins, labels=group_names)
sorted_final_school_data.head(3)
```

```python
#Now group based on Spending Per Student. Then cut out columns other than the target columns,
spending_school_data = sorted_final_school_data.groupby(['Spending Per Student'])
spending_final_summary = spending_school_data[['Math Average', 'Reading Average', '% Pass Math', '% Pass Reading', '% Passing']]
spending_final_summary.mean()
```
